```python
# 1. Write a program that takes a word as an input and print the number of vowels in the word.

word = input("Please type in the word: ").lower()

vowel_count = 0
for i in word:
    if i in ('a', 'e', 'i', 'o', 'u'):
        vowel_count += 1
    else:
        vowel_count += 0
print(f'number of vowel in this word: {vowel_count}')
```
