```python
##NUMPY
# Question: 1) Define two custom numpy arrays, say A and B. Generate two new numpy arrays by stacking A and B vertically and horizontally.
import numpy as np

# From creating a custom array from list
A = np.array([1,2,3,4,5])
B = np.array([4,5,6,7,8])

vertical_stack = np.vstack([A,B])
horizontal_stack = np.hstack([A,B])
print(horizontal_stack)
print(vertical_stack)
```

```python
## NUMPY
# Question: 2) Find common elements between A and B. [Hint : Intersection of two sets]
common_element = np.intersect1d(A,B)
print(common_element)
```

```python
## NUMPY
# Question: 3) Extract all numbers from A which are within a specific range. eg between 5 and 10. [Hint: np.where() might be useful or boolean masks]
within_range = np.where((A >= 5) & (A < 10))
print("Index: ", within_range)
print("Value: ", A[within_range])
```

```python
## NUMPY
# Question: 4) Filter the rows of iris_2d that has petallength (3rd column) > 1.5 and sepallength (1st column) < 5.0
# url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
# iris_2d = np.genfromtxt(url, delimiter=',', dtype='float', usecols=[0,1,2,3])

import numpy as np
from sklearn.datasets import load_iris

# load numeric features only (150 x 4)
iris = load_iris()
iris_2d = iris.data  # columns: [sepal_length, sepal_width, petal_length, petal_width]

# Filter: petal_length (col 2) > 1.5 and sepal_length (col 0) < 5.0
mask = (iris_2d[:, 2] > 1.5) & (iris_2d[:, 0] < 5.0)
filtered_rows = iris_2d[mask]

print(filtered_rows[:5])
print("Rows matched:", filtered_rows.shape[0])
```

```python
## PANDAS
# Question: 1) From df filter the 'Manufacturer', 'Model' and 'Type' for every 20th row starting from 1st (row 0).
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/selva86/datasets/master/Cars93_miss.csv')
print(df.head())
resulting_row = df.loc[::20, ['Manufacturer', 'Model', 'Type']]
print(resulting_row)
```

```python
## PANDAS
# Question: 2) Replace missing values in Min.Price and Max.Price columns with their respective mean.

min_mean = df['Min.Price'].mean()
max_mean = df['Max.Price'].mean()

# replace with NaN of max price and min price to the column mean
df['Min.Price'].fillna(min_mean, inplace = True)
df['Max.Price'].fillna(max_mean, inplace = True)

print(df[['Min.Price','Max.Price']].head())






```
